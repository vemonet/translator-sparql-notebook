```python
import json
gwas_url = 'https://translator.broadinstitute.org/genetics_data_provider/query'
query = """
{
  "message": {
    "query_graph": {
      "edges": [
        {
          "id": "e00",
          "source_id": "n00",
          "target_id": "n01",
          "type": "associated"
        }
      ],
      "nodes": [
        {
          "curie": "MONDO:0007455",
          "id": "n00",
          "type": "disease"
        },
        {
          "id": "n01",
          "type": "gene"
        }
      ]
    }
  }
}
"""
query = json.loads(query)
```

```python
import requests
gwas_response = requests.post(gwas_url, json=query).json()
gwas_response
```

```python
gene_ids = [node['id'] for node in gwas_response['knowledge_graph']['nodes'] if node['type']=='gene']
len(gene_ids)
```

```python
gene_ids[0:10]
```

```python
symbols = []
for i in range(len(gene_ids)):
    response = requests.get('https://mygene.info/v3/gene/'+gene_ids[i][9:]).json()
    print(response['symbol'])
    symbols.append(response['symbol'])
symbols = ";".join(symbols)
symbols
```

```python
base_url = 'https://translator.broadinstitute.org/molecular_data_provider'
```

```python
controls = [{'name':'gene symbols', 'value':symbols}]
query = {'name':'HGNC gene-list producer', 'controls':controls}
gene_list = requests.post(base_url+'/transform', json=query).json()
gene_list
```

```python
requests.get(base_url+'/gene/list/'+gene_list['id']).json()
```

```python
controls = []
query = {'name':'DrugBank inhibitors transformer','collection_id':gene_list['id'], 'controls':controls}
compound_list = requests.post(base_url+'/transform', json=query).json()
compound_list 
```

```python
controls = [{'name':'score threshold', 'value':'99'},{'name':'maximum number', 'value':'2'}]
query = {'name':'CMAP compound-to-compound expander','collection_id':compound_list['id'], 'controls':controls}
cmap_comex_list = requests.post(base_url+'/transform', json=query).json()
cmap_comex_list
```

```python
controls = [{'name':'maximum FDR', 'value':'0.2'},{'name':'disease context', 'value':'pan-cancer (all lines)'},{'name': 'maximum number', 'value':'3'}]
query = {'name':'CTRP compound-list expander','collection_id':compound_list['id'], 'controls':controls}
ctrp_comex_list = requests.post(base_url+'/transform', json=query).json()
ctrp_comex_list
```

```python
query = {'operation':'union', 'collection_ids':[cmap_comex_list['id'], ctrp_comex_list['id']]}

gene_similarity_list = requests.post(base_url+'/aggregate', json=query).json()
gene_similarity_list
```

```python
query = {'operation':'difference', 'collection_ids':[gene_similarity_list['id'], compound_list['id']]}

diff_list = requests.post(base_url+'/aggregate', json=query).json()
diff_list
```
